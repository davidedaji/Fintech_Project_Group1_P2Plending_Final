```python
#logit

import pandas as pd
import statsmodels.api as sm
import numpy as np

#import dataset
filename = 'C:/Users/Utente/Downloads/New_FinTech_input_data_10kshort1617.xlsx' 

selected_data1 =  pd.read_excel(filename, sheet_name=None, engine='openpyxl')
selected_data = pd.concat(selected_data1, ignore_index=True)


#create dummies for credit grade
selected_data['Credit Grade B']=0
selected_data.loc[(selected_data['grade']=='B'), 'Credit Grade B'] = 1
selected_data['Credit Grade C']=0
selected_data.loc[(selected_data['grade']=='C'), 'Credit Grade C'] = 1
selected_data['Credit Grade D']=0
selected_data.loc[(selected_data['grade']=='D'), 'Credit Grade D'] = 1
selected_data['Credit Grade E']=0
selected_data.loc[(selected_data['grade']=='E'), 'Credit Grade E'] = 1
selected_data['Credit Grade F']=0
selected_data.loc[(selected_data['grade']=='F'), 'Credit Grade F'] = 1
selected_data['Credit Grade G']=0
selected_data.loc[(selected_data['grade']=='G'), 'Credit Grade G'] = 1
#create dummies for home ownership
selected_data['home']=0
selected_data.loc[(selected_data['home_ownership']=='OWN'), 'home'] = 1
#create dummies for Fico score
selected_data['FICO 6']=0
selected_data.loc[(selected_data['fico_range_low'] >=780), 'FICO 6'] = 1
selected_data['FICO 5']=0
selected_data.loc[(selected_data['fico_range_low'] >=750) & (selected_data['fico_range_low'] <=779), 'FICO 5'] = 1
selected_data['FICO 4']=0
selected_data.loc[(selected_data['fico_range_low'] >=714) & (selected_data['fico_range_low'] <=749), 'FICO 4'] = 1
selected_data['FICO 3']=0
selected_data.loc[(selected_data['fico_range_low'] >=679) & (selected_data['fico_range_low'] <=713), 'FICO 3'] = 1
selected_data['FICO 2']=0
selected_data.loc[(selected_data['fico_range_low'] >=660) & (selected_data['fico_range_low'] <=678), 'FICO 2'] = 1
selected_data['FICO 1']=0
selected_data.loc[(selected_data['fico_range_low'] >=640) & (selected_data['fico_range_low'] <=659), 'FICO 1'] = 1
#create default
selected_data['default']=0
selected_data.loc[(selected_data['loan_status'] == 'Charged Off') | (selected_data['loan_status'] == 'Default'), 'default'] = 1
#log transformation
selected_data['log_annual_inc']=0
selected_data.loc[(selected_data['annual_inc'] != 0)  , 'log_annual_inc'] = np.log(selected_data['annual_inc'])
selected_data['log_dti']=0
selected_data.loc[(selected_data['dti'] != 0)  , 'log_dti'] = np.log(selected_data['dti'])
selected_data['log_revol_bal']=0
selected_data.loc[(selected_data['revol_bal'] != 0)  , 'log_revol_bal'] = np.log(selected_data['revol_bal'])
selected_data['log_installment']=np.log(selected_data['installment'])
selected_data['log_funded_amnt']=np.log(selected_data['funded_amnt'])


#change here variables of interest
Y = selected_data['default']
indep_var = ['Credit Grade B','Credit Grade C', 'Credit Grade D', 'Credit Grade E', 'Credit Grade F', 'Credit Grade G', 
             'FICO 2', 'FICO 3', 'FICO 4', 'FICO 5', 
             'log_annual_inc', 'log_dti', 'log_funded_amnt', 'int_rate', 'emp', 'log_installment', 'log_revol_bal']

#run the model
X = sm.tools.add_constant(selected_data[indep_var])
logit_mod = sm.Logit(Y, X)
logit_res = logit_mod.fit(cov_type='HC1')


print(logit_res.summary())
print (' Finished analyzing data  ')
print (' READY! ')
print(logit_res.get_margeff().summary()) 
```
