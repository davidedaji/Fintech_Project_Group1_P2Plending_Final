```python


# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces

fig.add_trace(
    go.Scatter(x=All_Fintech_Stat['issue_d'], y=All_Fintech_Stat['Average'], name="Average interest rate", line = dict(color='BLUE', width=3.5, )),
    secondary_y=False, 
)

fig.add_trace(
    go.Scatter(x=All_Fintech_Stat['issue_d'], y=All_Fintech_Stat['5% percentile'], name="5th percentile", line = dict(color='green', width=2,dash='dash' )),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=All_Fintech_Stat['issue_d'], y=All_Fintech_Stat['95% percentile'], name="95th percentile", line = dict(color='red', width=2, dash='dash')),
    secondary_y=False,
)
fig.add_trace(
    go.Scatter(x=All_Fintech_Stat['issue_d'], y=All_Fintech_Stat['US_GDP'], name="Real US GDP growth %", line = dict(color='black', width=2, dash='dash')),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="P2P lending Interest rate over time vs US real GDP growth"
)

# Set x-axis title
fig.update_xaxes(title_text="Months")

fig.update_yaxes(title_text="<b>Interest Rate and Saving ratio</b> ", secondary_y=False)
fig.update_yaxes(title_text=" % <b>GDP Growth</b> ", secondary_y=True)
fig.update_layout(width=1100, height=600)
fig.update_yaxes(tick0=0.0000, dtick=0.01, secondary_y =False)
fig.show()
fig.write_image("fig6.svg")
```
